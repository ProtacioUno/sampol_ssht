```python
import pandas as pd  
import numpy as np
import matplotlib.pyplot as plt
import math 
# import 
```

```python
df_ho = pd.read_excel('sampol.xlsx' , sheet_name='Sheet2')
df_aw = pd.read_excel('sampol.xlsx' , sheet_name='Sheet1') 
df_prev = pd.read_excel('combine_01.xlsx' , sheet_name='Sheet1')  
# display(df_prev)


def mergy(df_home , df_away, df_previus): 
    df_home = df_home.rename(columns=lambda x: x.replace(' ', '_'))
    df_away = df_away.rename(columns=lambda x: x.replace(' ', '_'))
    df_home['DATE'] = pd.to_datetime(df_home['DATE'])
    df_away['DATE'] = pd.to_datetime(df_away['DATE'])

    df_home
    df_home['row_num'] = df_home.groupby('LOAN_ID') .cumcount() + 1
    df_away['row_num'] = df_away.groupby('LOAN_ID') .cumcount() + 1
    df_home['loan_row'] = df_home['LOAN_ID'].astype(str) + "_" + df_home['row_num'].astype(str)
    df_away['loan_row'] = df_away['LOAN_ID'].astype(str) + "_" + df_away['row_num'].astype(str)
    df_inner = pd.merge(df_home, df_away , on=['LOAN_ID' , 'row_num'] , how='inner') 
    df_inner['LOAN-row'] = df_inner['LOAN_ID'].astype(str) + "_" + df_inner['row_num'].astype(str)
    loan_id =  df_inner['loan_row_y'].tolist() 
    # print(loan_id)
    # must_merge = df_away.query("LOAN_ID != @loan_id") 
    must_merge = df_away.query("loan_row != @loan_id") 
    # display(must_merge)
    combined_df = pd.concat([df_home, must_merge], ignore_index=True)
    
    if not df_previus.empty:
        df_previus = df_previus[['loan_row' , 'collector']]
        combined_df_merge = pd.merge(combined_df, df_previus , on='loan_row' , how='left')
        display(combined_df_merge)
        combined_df_merge.to_excel("combine.xlsx" , index=False, engine='openpyxl')
        return combined_df_merge
    else : 
        return combined_df  
        
    display(combined_df) 

    mergy(df_ho , df_aw , df_prev)
```
